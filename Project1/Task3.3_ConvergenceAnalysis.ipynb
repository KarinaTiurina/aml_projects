## Task 3.3

Convergence analysis: check how the value of log-likelihood function depends on the number of itera
tions for 3 above algorithms. Convergence analysis should be performed on the train data.

```python
import numpy as np
import pandas as pd
from models import IRLS
from models import SGD
from models import ADAM
from sklearn.metrics import balanced_accuracy_score
from sklearn.model_selection import train_test_split
import datasets.read as data
from matplotlib import pyplot as plt
import seaborn as sns
from datasets.read import read_all_datasets
import json

datasets = read_all_datasets('datasets/');
```

```python
# Print shapes of datasets
for dataset_name, dataset in datasets.items():
    X, y = dataset
    if dataset_name == 'Web_Page_Phishing' or dataset_name == 'Airline_Passenger_Satisfaction':
        fraction = int(X.shape[0]*0.5)
        X = X[:fraction]
        y = y[:fraction]
    print(f"{dataset_name}: {X.shape}, {y.shape}")
```

## IWLS

```python
history = {}

for dataset_name, dataset in datasets.items():
  X, y = dataset
  if dataset_name == 'Web_Page_Phishing' or dataset_name == 'Airline_Passenger_Satisfaction':
    fraction = int(X.shape[0]*0.5)
    X = X[:fraction]
    y = y[:fraction]
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)
  irls = IRLS()
  irls.fit(X_train, y_train)
  history[dataset_name] = irls.get_params()['loss_history']
  y_pred = irls.predict(X_test)
  print(f'{dataset_name}: {balanced_accuracy_score(y_test, y_pred)*100:.2f}%')

# Save history
file_path = "task3.3_results/irls.json"
with open(file_path, 'w') as json_file:
    json.dump(history, json_file)

# Plot loss
for dataset_name, data in history.items():
  plt.plot(data, label=dataset_name)

plt.legend()
plt.title('IWLS loss per iteration')
plt.xlabel('Iteration')
plt.ylabel('Loss');
```
